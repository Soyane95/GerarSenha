```python
import pandas as pd

dados = {
    "Nome": ["Ana", "Carlos", "João", "Ana", "Mariana", "Pedro", "Lucas"],
    "Idade": [25, None, 28, 25, None, 23, None],
    "Cidade": ["Rio", "São Paulo", None, "Rio", "Belo Horizonte", "Rio", "Curitiba"]
}

df = pd.DataFrame(dados)
df
```

```python
# remover duplicados
df = df.drop_duplicates()

# preencher valores nulos
df["Idade"] = df["Idade"].fillna(0)
df["Cidade"] = df["Cidade"].fillna("Não informado")

df
```

```python
# média de idade
media = df["Idade"].mean()

# quantidade por cidade
cidade = df["Cidade"].value_counts()

print("Média de idade:", media)
print("\nPessoas por cidade:\n", cidade)
```
